# 반복문

```python
count = 0
while count < 10:
    count = count +1
    print("나무를 %d번 찍었습니다." % count)
    if count == 10:
        print("나무 넘어갑니다.")
```

# 무한루프

```python
while True:
    print("반복을 계속 할까요? [예/아니오] :")
    answer = input()
    
    if answer == '예':
        print("반복을 계속합니다.")
    elif answer == '아니오':
        break
    else:
        print("정상적인 답변이 아닙니다.")
```

# 예제
#"총 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 보여 주시오."
#90, 25, 67, 45, 80점


```python
marks = [90, 25, 67, 45, 80]
index = 0 
for mark in marks: 
    index = index +1 
    if mark >= 60: 
        print("%d번 학생은 합격입니다." % index)
    else: 
        print("%d번 학생은 불합격입니다." % index)
```

# 예제
#"총 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 보여 주시오."
#90, 25, 67, 45, 80점


```python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60: 
        continue
    print("%d번 학생 축하합니다. 합격입니다." % (number+1))
```
